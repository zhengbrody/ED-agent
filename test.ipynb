```python

import json
from EDAgentLLM import EDAgentLLM
from ToolExecutor import ToolExecutor
from EDAgent import ReActAgent

def main():
    llm_client = EDAgentLLM()  #LLM API wrapper
    toolExecutor = ToolExecutor() # Tools


    agent = ReActAgent(llm_client, toolExecutor, max_steps=10)


    # 2. 测试案例
    questions = [
        # case 1
        "Load the IEEE118 case and solve the economic dispatch problem.",

        # case 2
        """
        what if we increase the load by 10% and solve again.
        """,

        # case 3
        """
        Load IEEE14 case. Shut down generator 2 (set status to 0).
        Then solve the economic dispatch for the remaining generators.
        """,
    ]

    # 3. 运行测试
    for i, question in enumerate(questions, 1):
        print(f"\n\n{'#'*70}")
        print(f"# case {i}")
        print(f"{'#'*70}")
        print(f"question: {question}\n")

        result = agent.run(question)

        if result:
            print(f"\n✅ generate success!!!")
            if isinstance(result, dict):
                print(json.dumps(result, indent=2, ensure_ascii=False))


if __name__ == "__main__":
    main()
```
